## Part II. Authentic chemical library annotation



```python
import json
import pandas as pd
import numpy as np
from asari.tools.match_features import list_match_lcms_features
```

```python
def list_cpdlib_match(features, cpdlib, mz_ppm=5, rt_tolerance=10):
    '''
    return all matches within m/z ppm and RT tolerance from cpdlib for each feature. 
    '''
    for f in features:
        f['id'] = f['id_number']
    for c in cpdlib:
        c['id'] = c['cpd']
    matched = list_match_lcms_features(features, cpdlib, mz_ppm=mz_ppm, rt_tolerance=rt_tolerance)
    return matched
```

## Part III. Linking MS1 and MS2 annotations to empirical compounds

Other annotation sources can be integrated into empCpds similarly.

```python
# read empCpds from json
empCpds_json = '/Users/lish/li.proj/staged_dev_mcg3/mzML/processed_MS1_asari_project_112712271/Annotated_empiricalCompounds.json'
empCpds = json.load(open(empCpds_json))
len(empCpds)
```

```python
# Make a dict to link features to empCpds, 
# MS2 search results are in `results` from above
feature_to_empCpd = {}
for emp_id, emp in empCpds.items():
    for f in emp['MS1_pseudo_Spectra']:
        feature_to_empCpd[f['id_number']] = emp_id
        
# MS1 to authLib, result in ms1_matched from above
# Make a dict to link features to authLib compounds
feature_dict = {f['id_number']: f for f in LCMS_features}
compound_dict = {c['cpd']: c for c in authLib['list_features']}

# Optional here, but additional cpd info may be in master compound table
# cpdTable_output = '/Users/lish/li.proj/staged_dev_mcg3/compoundTable_LiLab_library_11_06_2024.json'
# compound_table = json.load(open(cpdTable_output))
# len(compound_table), compound_table[0]
# compound_dict = {c['identifier']: c for c in compound_table}
```

```python
compound_dict['HMDB0000030']
```

```python
# Now chain annotations into empCpds
# We will select from authLib match the best RT matched compound; not the best,
# to improve later
# also only use those features that are primary ion 'M0,M+H+'
def find_best_rt_match(feat, cpd_ids, compound_dict):
    feat_rt = feat['rtime']
    best_cpd = None
    best_rt_diff = None
    for cpd_id in cpd_ids:
        cpd = compound_dict[cpd_id]
        cpd_rt = cpd.get('rtime', None)
        if cpd_rt is None:
            continue
        rt_diff = abs(cpd_rt - feat_rt)
        if (best_rt_diff is None) or (rt_diff < best_rt_diff):
            best_rt_diff = rt_diff
            best_cpd = cpd
    return best_cpd

def force_primary_ion(feat_id, epd, primary_ion='M0,M+H+'):
    '''
    Force the primary ion in the annotation to be the given one.
    '''
    anno_dict = {f['id_number']: f.get('ion_relation', None) for f in epd['MS1_pseudo_Spectra']}
    if anno_dict.get(feat_id, None) == primary_ion:
        return True
    else:
        return False
    
for feat_id, cpd_ids in ms1_matched.items():
    emp_id = feature_to_empCpd.get(feat_id, None)
    if emp_id is None or not force_primary_ion(feat_id, empCpds[emp_id], primary_ion='M0,M+H+'):
        continue
    emp = empCpds[emp_id]
    feat = feature_dict[feat_id]
    best_cpd = find_best_rt_match(feat, cpd_ids, compound_dict)
    if best_cpd is not None:
        if 'annotation' not in emp:
            emp['annotation'] = {}
        if "authLib_Li_Lab_202410" not in emp['annotation']:
            emp['annotation']['authLib_Li_Lab_202410'] = []
        emp['annotation']['authLib_Li_Lab_202410'].append(
            {feat_id: best_cpd}
        )
```

```python
# go on to MS2 search results
for feat_id, ms2_results in results.items():
    emp_id = feature_to_empCpd.get(feat_id, None)
    if emp_id is None:
        continue
    emp = empCpds[emp_id]
    if 'annotation' not in emp:
        emp['annotation'] = {}
    if "MS2_entropy_search" not in emp['annotation']:
        emp['annotation']['MS2_entropy_search'] = []
    emp['annotation']['MS2_entropy_search'].append(
        {feat_id: ms2_results}
    )
```

## Part IV. Differential metabolomic features by rotenone treatment

Data analysis uses preferred feature table. 

Annotation is based on full table.

```python
import numpy as np
from scipy.stats import ttest_ind
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ms1_table = "/Users/lish/li.proj/staged_dev_mcg3/mzML/processed_MS1_asari_project_112712271/preferred_Feature_table.tsv"
df_features = pd.read_csv(ms1_table, sep='\t')
df_features.head()
```

```python
df_features.columns[10:]
```

```python
stats_result = []
for idx, row in df_features.iterrows():
    feat_id = row['id_number']
    intensities = list(row.values[11:])
    intensities = np.log2(np.array(intensities) + 1)  # log2 transform
    group1 = intensities[:3]
    group2 = intensities[3:6]
    t_stat, p_value = ttest_ind(group1, group2, nan_policy='omit')
    stats_result.append({
        'feat_id': feat_id,
        'mz': row['mz'],
        'rt': row['rtime'],
        't_stat': t_stat,
        'p_value': p_value,
    })
```

```python
stats_result.sort(key=lambda x: x['p_value'])
```

```python
s = '\t'.join(['m/z', 'retention_time', 'p-value', 't-score', 'feature_id']) + '\n'
for x in stats_result:
    s += '\t'.join([str(x['mz']), str(x['rt']), str(x['p_value']), 
                    str(x['t_stat']), str(x['feat_id'])]) + '\n'
with open('ineurons_ttest_1127.tsv', 'w') as f:
    f.write(s)
```
